```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def scaled_sigmoid(t, tmax=512):
    """
    Sigmoid function that is 0 at t=0 and 1 at t=512.
    
    Parameters:
    t (float or np.ndarray): Input value or array of values.
    
    Returns:
    float or np.ndarray: Sigmoid output.
    """
    # Calculate the scaling factor
    t0 = tmax / 2
    epsilon = 1e-4
    k = np.log((1 - epsilon) / epsilon) / (t0 - 0)
    
    # Shift and scale the input to fit the desired range
    return 1 / (1 + np.exp(-k * (t - t0)))
```

```python
# plot
tmax = 512
t = np.linspace(0, 512, 500)
plt.plot(t, scaled_sigmoid(t, tmax=512));
```
