```python
import optax
import numpy as np
import pandas as pd
```

```python
n_train_samples = 162770
bs = 256
n_epochs = 256
n_updates_epoch = n_train_samples // bs
n_updates = n_epochs * n_updates_epoch
t = np.arange(1,n_updates) 
```

```python
tx = optax.schedules.cosine_decay_schedule(1e-4, n_updates, alpha=0.0)
pd.Series(tx(t)).plot()
```

```python
tx = optax.schedules.warmup_cosine_decay_schedule(1e-5, 1e-2, 25, n_updates)
pd.Series(tx(t)).plot()
```

```python
# init_value: float,
# peak_value: float,
# warmup_steps: int,
# transition_steps: int,
# decay_rate: float,
# transition_begin: int = 0,
# staircase: bool = False,
# end_value: Optional[float] = None

# init_value: 0.0
#   peak_value: ${training.lr}
#   warmup_steps: 1000
#   transition_steps: 100000
#   decay_rate: 0.5

tx = optax.schedules.warmup_exponential_decay_schedule(0, 1e-4, 1000, 100000, 0.30)
pd.Series(tx(t)).plot()
```
